
# Language Concepts and Linguistic Phenomena

## 1. Evolutionary Concepts

**Smilodon and Thylacosmilus:** Examples of **Parallel/Convergent Evolution** where similar structures evolved independently in species due to similar environmental pressures.

**Example:**

- Environment: Water
- Feature: **Interdigital Webbing** seen in ducks, swans, frogs, and penguins for efficient swimming.

---

## 2. Distributional Hypothesis

**Similar Context ‚Üí Similar Meaning**

- Sentence 1: "The cat is sleeping on the **sofa**."
- Sentence 2: "The cat is sleeping on the **couch**."
  **Sofa** and **Couch** convey similar meanings due to their shared context.

**Different Context ‚Üí Different Meaning**

- Sentence 1: "I ate an **apple** for breakfast." (*Fruit*)
- Sentence 2: "I bought an **Apple** iPhone." (*Company*)
  **Apple** holds distinct meanings in different contexts.

## 3. Word Embeddings

**Numerical representations of word meanings.**

**Types of Embeddings:**

- **Static Embeddings**
  - Example: Word2Vec, GloVe
  - **Appleüçé** ‚Üí [1 2 3]
  - **Appleüì±** ‚Üí [1 2 3] (Same vector regardless of context)

- **Dynamic (Contextualized) Embeddings**
  - Example: BERT
  - **Appleüçé** ‚Üí [1 2 3]
  - **Appleüì±** ‚Üí [4 5 6] (Changes based on context)

---

## 4. Lexical Relations in Linguistics

- **Hypernym:** General term (e.g., Animal)
- **Hyponym:** Specific example (e.g., Tiger)
- **Co-Hyponyms:** Terms at the same hierarchical level (e.g., Tiger, Lion, Cheetah)

- **Meronymy:** Part of something (e.g., Wheel ‚Üí Car)
- **Holonymy:** Whole of something (e.g., Car ‚Üí Wheel, Engine)

---

## 5. Topic Modeling & Semantic Fields

**Topic Modeling Model (TMM):** Groups words related to a specific subject.

- **TMM(Movie, Actor, IMDB Rating, Audience, Snacks) ‚Üí Theatre**
- Example Model: **Latent Dirichlet Allocation (LDA)**

**Semantic Fields:** Clusters of related words within a topic.

- Example Field: **Theatre**
  - Words: **Movie, Actor, IMDB Rating, Audience, Snacks**
